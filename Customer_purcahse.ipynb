```python
import json
fd = open('Records.json', 'r')
r = fd.read()
fd.close()

Record = json.loads(r)

fd = open('Sales.json', 'r')
s = fd.read()
fd.close()

sales = json.loads(s)
```

```python
from datetime import datetime
n=-1
while n!=0:
    print("Welcome to the Garry's Groecry store!")
    print("0 - Exit \n1 - Fruits \n2 - Vegetables \n3 - flour \n4 - Condiments-spices \n5 - Fish-Shelfish \n6 - Baverages\n")
    n=int(input("Enter the No. as mentionned above what you want to buy from our store : "))
    if n!=0:
        cust_name=str(input("Enter Your Name : "))
        cust_mob_no=int(input("Enter the Mobile Number : "))
        cust_city=str(input("Enter Your City Name : "))
    #Fruits
    if n==1:
        print("\n<--*--*--*--*--*--*Available products are*--*--*--*--*--*--*-->\n")
        print('ID   -  Prroduct')
        for i in Record['Fruits']:
            print(i,' - ',Record['Fruits'][i]['Name'])
        a=str(input("\nEnter the id of prodcut which you want to buy......"))
        q=int(int(input("\nEnter the quantity how much you want :")))
        if q<Record['Fruits'][a]['quantity']:
            for i in Record['Fruits']:
                if i==a:
                    print("\n*--*--*--*--*--*--*--*--*Your bill is this*--*--*--*--*--*--*--*--*")
                    print('     Product ID - ',i)
                    print('     Product Name - ',Record['Fruits'][i]['Name'])
                    print('     Product price - '+str(Record['Fruits'][i]['price'])+' /',Record['Fruits'][i]['unit'])
                    print('     Total Amount - '+ str(Record['Fruits'][i]['price']*q))
                    print("*--*--*--*--*--*--*--*--*PLASE VISIT AGIAN*--*--*--*--*--*--*--*--*\n")
                    Record['Fruits'][i]['quantity']=Record['Fruits'][i]['quantity']-q
                    now = datetime.now()
                    current_time = now.strftime("%H:%M:%S")
                    sales[len(sales)+1]={'Name':cust_name,'contact':cust_mob_no,'Address':cust_city,'product':Record['Fruits'][i]['Name'],'amount':Record['Fruits'][i]['price']*q,'time':current_time}
        else:
            print(f"\nWe are Sorry your quantity is too large we only have {Record['Fruits'][a]['quantity']}\n")
    #Vegetables
    elif n==2:
        print("\n<--*--*--*--*--*--*Available products are*--*--*--*--*--*--*-->\n")
        print('ID   -  Prroduct')
        for i in Record['Vegetables']:
            print(i,' - ',Record['Vegetables'][i]['Name'])
        a=str(input("\nEnter the id of prodcut which you want to buy......"))
        q=int(int(input("\nEnter the quantity how much you want :")))
        if q<Record['Vegetables'][a]['quantity']:
            for i in Record['Vegetables']:
                if i==a:
                    print("\n*--*--*--*--*--*--*--*--*Your bill is this*--*--*--*--*--*--*--*--*")
                    print('     Product ID - ',i)
                    print('     Product Name - ',Record['Vegetables'][i]['Name'])
                    print('     Product price - '+str(Record['Vegetables'][i]['price'])+' /',Record['Vegetables'][i]['unit'])
                    print('     Total Amount - '+ str(Record['Vegetables'][i]['price']*q))
                    print("*--*--*--*--*--*--*--*--*PLASE VISIT AGIAN*--*--*--*--*--*--*--*--*\n")
                    Record['Vegetables'][i]['quantity']=Record['Vegetables'][i]['quantity']-q
                    now = datetime.now()
                    current_time = now.strftime("%H:%M:%S")
                    sales[len(sales)+1]={'Name':cust_name,'contact':cust_mob_no,'Address':cust_city,'product':Record['Vegetables'][i]['Name'],'amount':Record['Vegetables'][i]['price']*q,'time':current_time}
        else:
            print(f"\nWe are Sorry your quantity is too large we only have {Record['Vegetables'][a]['quantity']}\n")
    #flour
    elif n==3:
        print("\n<--*--*--*--*--*--*Available products are*--*--*--*--*--*--*-->\n")
        print('ID   -  Prroduct')
        for i in Record['flour']:
            print(i,' - ',Record['flour'][i]['Name'])
        a=str(input("\nEnter the id of prodcut which you want to buy......"))
        q=int(int(input("\nEnter the quantity how much you want :")))
        if q<Record['flour'][a]['quantity']:
            for i in Record['flour']:
                if i==a:
                    print("\n*--*--*--*--*--*--*--*--*Your bill is this*--*--*--*--*--*--*--*--*")
                    print('     Product ID - ',i)
                    print('     Product Name - ',Record['flour'][i]['Name'])
                    print('     Product price - '+str(Record['flour'][i]['price'])+' /',Record['flour'][i]['unit'])
                    print('     Total Amount - '+ str(Record['flour'][i]['price']*q))
                    print("*--*--*--*--*--*--*--*--*PLASE VISIT AGIAN*--*--*--*--*--*--*--*--*\n")
                    Record['flour'][i]['quantity']=Record['flour'][i]['quantity']-q
                    now = datetime.now()
                    current_time = now.strftime("%H:%M:%S")
                    sales[len(sales)+1]={'Name':cust_name,'contact':cust_mob_no,'Address':cust_city,'product':Record['flour'][i]['Name'],'amount':Record['flour'][i]['price']*q,'time':current_time}
        else:
            print(f"\nWe are Sorry your quantity is too large we only have {Record['flour'][a]['quantity']}\n")
    #Condiments-spices
    elif n==4:
        print("\n<--*--*--*--*--*--*Available products are*--*--*--*--*--*--*-->\n")
        print('ID   -  Prroduct')
        for i in Record['Condiments-spices']:
            print(i,' - ',Record['Condiments-spices'][i]['Name'])
        a=str(input("\nEnter the id of prodcut which you want to buy......"))
        q=int(int(input("\nEnter the quantity how much you want :")))
        if q<Record['Condiments-spices'][a]['quantity']:
            for i in Record['Condiments-spices']:
                if i==a:
                    print("\n*--*--*--*--*--*--*--*--*Your bill is this*--*--*--*--*--*--*--*--*")
                    print('     Product ID - ',i)
                    print('     Product Name - ',Record['Condiments-spices'][i]['Name'])
                    print('     Product price - '+str(Record['Condiments-spices'][i]['price'])+' /',Record['Condiments-spices'][i]['unit'])
                    print('     Total Amount - '+ str(Record['Condiments-spices'][i]['price']*q))
                    print("*--*--*--*--*--*--*--*--*PLASE VISIT AGIAN*--*--*--*--*--*--*--*--*\n")
                    Record['Condiments-spices'][i]['quantity']=Record['Condiments-spices'][i]['quantity']-q
                    now = datetime.now()
                    current_time = now.strftime("%H:%M:%S")
                    sales[len(sales)+1]={'Name':cust_name,'contact':cust_mob_no,'Address':cust_city,'product':Record['Condiments-spices'][i]['Name'],'amount':Record['Condiments-spices'][i]['price']*q,'time':current_time}
        else:
            print(f"\nWe are Sorry your quantity is too large we only have {Record['Condiments-spices'][a]['quantity']}\n")  
    #Fish-shelfish
    elif n==5:
        print("\n<--*--*--*--*--*--*Available products are*--*--*--*--*--*--*-->\n")
        print('ID   -  Prroduct')
        for i in Record['Fish-Shelfish']:
            print(i,' - ',Record['Fish-Shelfish'][i]['Name'])
        a=str(input("\nEnter the id of prodcut which you want to buy......"))
        q=int(int(input("\nEnter the quantity how much you want :")))
        if q<Record['Fish-Shelfish'][a]['quantity']:
            for i in Record['Fish-Shelfish']:
                if i==a:
                    print("\n*--*--*--*--*--*--*--*--*Your bill is this*--*--*--*--*--*--*--*--*")
                    print('     Product ID - ',i)
                    print('     Product Name - ',Record['Fish-Shelfish'][i]['Name'])
                    print('     Product price - '+str(Record['Fish-Shelfish'][i]['price'])+' /',Record['Fish-Shelfish'][i]['unit'])
                    print('     Total Amount - '+ str(Record['Fish-Shelfish'][i]['price']*q))
                    print("*--*--*--*--*--*--*--*--*PLASE VISIT AGIAN*--*--*--*--*--*--*--*--*\n")
                    Record['Fish-Shelfish'][i]['quantity']=Record['Fish-Shelfish'][i]['quantity']-q
                    now = datetime.now()
                    current_time = now.strftime("%H:%M:%S")
                    sales[len(sales)+1]={'Name':cust_name,'contact':cust_mob_no,'Address':cust_city,'product':Record['Fish-Shelfish'][i]['Name'],'amount':Record['Fish-Shelfish'][i]['price']*q,'time':current_time}
        else:
            print(f"\nWe are Sorry your quantity is too large we only have {Record['Fish-Shelfish'][a]['quantity']}\n")
    #Baverages
    elif n==6:
        print("\n<--*--*--*--*--*--*Available products are*--*--*--*--*--*--*-->\n")
        print('ID   -  Prroduct')
        for i in Record['Baverages']:
            print(i,' - ',Record['Baverages'][i]['Name'])
        a=str(input("\nEnter the id of prodcut which you want to buy......"))
        q=int(int(input("\nEnter the quantity how much you want :")))
        if q<Record['Baverages'][a]['quantity']:
            for i in Record['Baverages']:
                if i==a:
                    print("\n*--*--*--*--*--*--*--*--*Your bill is this*--*--*--*--*--*--*--*--*")
                    print('     Product ID - ',i)
                    print('     Product Name - ',Record['Baverages'][i]['Name'])
                    print('     Product price - '+str(Record['Baverages'][i]['price'])+' /',Record['Baverages'][i]['unit'])
                    print('     Total Amount - '+ str(Record['Baverages'][i]['price']*q))
                    print("*--*--*--*--*--*--*--*--*PLASE VISIT AGIAN*--*--*--*--*--*--*--*--*\n")
                    Record['Baverages'][i]['quantity']=Record['Baverages'][i]['quantity']-q
                    now = datetime.now()
                    current_time = now.strftime("%H:%M:%S")
                    sales[len(sales)+1]={'Name':cust_name,'contact':cust_mob_no,'Address':cust_city,'product':Record['Baverages'][i]['Name'],'amount':Record['Baverages'][i]['price']*q,'time':current_time}
        else:
            print(f"\nWe are Sorry your quantity is too large we only have {Record['flour'][a]['quantity']}\n")
```
